1. **Write the three address code (TAC) of the following:**
  > a. Any if-statment of your choice

  **CODE**


```
{ 
  if (a<b)
   cout<<1;
  else
    cout<<0;
}
```

**TAC**

  1. if(a<b) goto 4
  2. t1 = 0
  3. goto 5
  4. t1 = 1
  5. end

  > b. Any for-loop of your choice


**CODE**

```
{ 
  for(i = 1; i<=10; i++)
    {
      a[i] = z * 5;                                       
    } 
}
```

**TAC**

  1. i = 1;
  2. L : t1 = z * 5
  3. t2 = &a
  4. t3 = sizeof(int)
  5. t4 = t3 * i
  6. t5 = t2 + t4
  7. t5 = t1
  8. i = i + 1
  9. if(i<=10) goto L
