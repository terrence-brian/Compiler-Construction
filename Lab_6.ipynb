**Introduction**  
The Frontend is made of three main processes. Lexical Analysis, Syntax Analysis, and semantic Analysis. These three processes are mostly combined into modern parsers as opposed to previous compilers that had different programs for each of the processes.

**Lexical Analyzer**  
The Lexical analyzer takes a string indicating a program and breaks it into tokens -- Tokenization. These tokens are obtained based on the rules of the programming Language and it's keywords. When tokenization is done, the Lexical analyzer returns the tokenized program.

**Parser**

> The Parser, also referred to as Syntax Analysis, takes the output of the Lexical Analysis and produces a syntax tree as output. This is done by taking all the token gotten from the Lexical Analyzer one by one and using Context-Free Grammar to construct the syntax tree. 

> This syntax tree, also referred to as parse tree or derivation tree, is constructed to check for ambiguity in the given grammar. The syntax tree follows the following rules:
* Any identifier is regarded an expression.
* Any number can be called an expression.
* Performing any operations in the given expression will always result in an expression.

**Semantic Analyzer**
> The Semantic Analyzer makes sure that declarations and statements of program are semantically correct. It checks whether the given program is semantically consistent with the language definition using a syntax tree and a symbol table. It collects type information and saves it in a syntax tree or a symbol table. This type information is then used by the compiler to generate intermediate code.
